```python
import threading
```

```python
def print_squares():
    for i in range(1, 11):
        print(f"Квадрат {i} = {i ** 2}")

def print_cubes():
    for i in range(1, 11):
        print(f"Куб {i} = {i ** 3}")
```

```python
t1 = threading.Thread(target=print_squares)
t2 = threading.Thread(target=print_cubes)

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
import time
```

```python
def print_numbers():
    for i in range(1, 11):
        print(i)
        time.sleep(1)
```

```python
threads = []
for _ in range(3): 
    t = threading.Thread(target=print_numbers)
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```
