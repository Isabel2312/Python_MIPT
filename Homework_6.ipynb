```python
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel

app = FastAPI()
```

```python
class Expression(BaseModel):
    expr: str
```

```python
current_expression = ""

@app.post("/set_expression")
def set_expression(expression: Expression):
    global current_expression
    current_expression = expression.expr
    return {"message": "Expression set"}

@app.get("/get_expression")
def get_expression():
    return {"expression": current_expression}

@app.post("/evaluate")
def evaluate():
    global current_expression
    try:
        result = eval(current_expression, {"__builtins__": None}, {})
        return {"result": result}
    except Exception as e:
        raise HTTPException(status_code=400, detail=f"Error: {e}")

@app.get("/add")
def add(a: float, b: float):
    return {"result": a + b}

@app.get("/subtract")
def subtract(a: float, b: float):
    return {"result": a - b}

@app.get("/multiply")
def multiply(a: float, b: float):
    return {"result": a * b}

@app.get("/divide")
def divide(a: float, b: float):
    if b == 0:
        raise HTTPException(status_code=400, detail="Division by zero")
    return {"result": a / b}
```
